```python
import seaborn as sns
sns.set(style="ticks")

# Load the example tips dataset
tips = sns.load_dataset("tips")

# Draw a nested boxplot to show bills by day and sex
sns.boxplot(x="day", y="total_bill", hue="sex", data=tips, palette="PRGn")
sns.despine(offset=10, trim=True)
```

```python
import pandas as pd
df_gfunctions_fwrite=pd.read_csv("files/gfunctions.fwrite.data", names=['fsize','func_elapsed','fwrite_elapsed','provider','lang'])
df_ibm_fwrite=pd.read_csv("files/ibm.fwrite.data", names=['fsize','func_elapsed','fwrite_elapsed','provider','lang'])
```

```python
new_df=pd.concat([df_gfunctions_fwrite, df_ibm_fwrite],ignore_index=True)
new_df
```

```python
for i, row in new_df.iterrows():
  new_df.set_value(i,'fwrite_elapsed_per_1k',row['fwrite_elapsed'] / (row['fsize']/1024.0))
  
new_df
```

```python
import numpy as np

new_df2=new_df.copy(deep=True)
```

```python
new_df3=new_df2[np.abs(new_df2.fwrite_elapsed_per_1k-new_df2.fwrite_elapsed_per_1k.mean())<=(0.5*new_df2.fwrite_elapsed_per_1k.std())]
new_df3
```
